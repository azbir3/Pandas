```python
#to create a dataset of unique players only
purchase_data_unique=purchase_data.drop_duplicates(["SN"])
purchase_data_unique

#create the bins 
bins = [0, 9, 14, 19, 24, 29, 34, 39, np.inf]
#create names for the bins
age_ranges=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#categorize players using the age bins
age_bucket=pd.cut(purchase_data_unique["Age"], bins,labels=age_ranges)
#age_bucket
Total_count_u=purchase_data_unique["Age"].groupby(age_bucket)
Total_count_u.count()
```
