```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#to group on player and count gender (i.e. drop duplicate player from the list)
group_SN=purchase_data.groupby('SN')['SN'].nunique()

#to calculated noted variables
purchaseID_sorted=purchase_data.groupby(["SN"])
purchase_count_sorted=purchaseID_sorted["Purchase ID"].count()
average_price_sorted=purchase_data.groupby('SN')["Price"].mean()
total_revenue_sorted=purchase_data.groupby('SN')["Price"].sum()
purch_perPerson=total_revenue_sorted/group_SN

#to format output table 
pd.options.display.float_format = '${:,.2f}'.format
# to create a summary for both count and percent of players
purch_analysis_SN=pd.DataFrame({
    "Purchase Count": purchase_count_sorted,
    "Average Purchase Price": average_price_sorted,
    "Total Purchase Value": total_revenue_sorted
})

purch_analysis_SN.reset_index().sort_values(by='Total Purchase Value', ascending=False).head()
```
